```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
length = 2048
data = np.loadtxt("output.txt",delimiter=',',dtype=float)
plt.figure(figsize = (64,64))
r = np.arange(0, 4 + 4/length, 4/length)
for i in range(length):    
    plt.scatter([r[i]] * length, data[i], c='b', s=0.5)
```
